```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# create the model
DTree = DecisionTreeClassifier(random_state = 42)
# model training
DTree.fit(X_train,y_train)
# prediction
DTree_prediction = DTree.predict(X_test)
# accuracy
DTree_acc = accuracy_score(y_test, DTree_prediction)
print(classification_report(y_test, DTree_prediction))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# create the model
LogReg = LogisticRegression(random_state = 42)
# model training
LogReg.fit(X_train, y_train)
# prediction
LogReg_pred = LogReg.predict(X_test)
# accuracy
LogReg_acc = accuracy_score(y_test, LogReg_pred)
print(classification_report(y_test, LogReg_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# create the model
KNN = KNeighborsClassifier()
# model training
KNN.fit(X_train, y_train)
# prediction
KNN_pred = KNN.predict(X_test)
# accuracy
KNN_acc = accuracy_score(y_test, KNN_pred)
print(classification_report(y_test, KNN_pred))
```

```python
from sklearn.svm import SVC, LinearSVC
```

```python
# create the model
SVM = SVC( random_state = 42)
# model training
SVM.fit(X_train, y_train)
# prediction
SVM_pred = SVM.predict(X_test)
# accuracy
SVM_acc = accuracy_score(y_test, SVM_pred)
print(classification_report(y_test, SVM_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_params = { 'n_neighbors' : [5,10,15,20],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
```

```python
gs = GridSearchCV(KNeighborsClassifier(), grid_params, verbose = 1, cv=3, n_jobs = -1)
```

```python
# fit the model on our train set
g_res = gs.fit(X_train, y_train)
```

```python
# find the best score
g_res.best_score_
```

```python
# get the hyperparameters with the best score
g_res.best_params_
```

```python
# use the best hyperparameters
knn = KNeighborsClassifier(n_neighbors = 20, weights = 'uniform',algorithm = 'brute',metric = 'manhattan')
knn.fit(X_train, y_train)
```

```python
from sklearn import decomposition, datasets
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
std_slc = StandardScaler()
```

```python
 pca = decomposition.PCA()
```

```python
dec_tree = tree.DecisionTreeClassifier()
```

```python
pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('dec_tree', dec_tree)])
```

```python
n_components = list(range(1,DATA.shape[1]+1,1))
```

```python
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8,10,12]
```

```python
parameters = dict(pca__n_components=n_components,
                      dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth)
```

```python
clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(X_train, y_train)
```

```python
print('Best Criterion:', clf_GS.best_estimator_.get_params()['dec_tree__criterion'])
print('Best max_depth:', clf_GS.best_estimator_.get_params()['dec_tree__max_depth'])
print('Best Number Of Components:', clf_GS.best_estimator_.get_params()['pca__n_components'])
print(); print(clf_GS.best_estimator_.get_params()['dec_tree'])
```
